```python
def get_dosimetric_fs(curve):
    dose = curve.relative_dose*100.0/curve.relative_dose[np.where(curve.axis == 0)[0][0]]
    # plt.figure()
    # plt.plot(curve.axis, dose)
    # plt.show()

    dose1 = dose[:len(dose)//2]
    dose2 = dose[len(dose)//2:]

    x_1 = np.interp(50, dose1, curve.axis[:len(dose)//2])
    x_2 = np.interp(50, dose2, curve.axis[len(dose)//2:])

    return x_2-x_1

def px_plot(x,y):
    # Create a figure
    import plotly.graph_objects as go
    fig = go.Figure()

    # Add a trace
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Line Plot'))

    # Update layout
    fig.update_layout(
        title='Line Plot using Plotly',
        xaxis=dict(
            title='X Axis (mm)',
            tickmode='linear',  # Use 'linear' to space ticks evenly
            dtick=5  # Set distance between ticks to 1 mm
        ),
        yaxis=dict(
            title='Y Axis',
            tickmode='linear',
            dtick=10  # Example for y-axis, set distance between ticks to 5
        )
    )

    # Show the figure
    fig.show()

for curve in calc_curves:
    if curve.coordinate != 'Z' and curve.particle == 0:
        print(curve.energy, curve.depth, curve.field_size)
        print(f'FS(50%) = {round(get_dosimetric_fs(curve)/10.0,1)} cm')
        px_plot(curve.axis, curve.relative_dose*100.0/curve.relative_dose[np.where(curve.axis == 0)[0][0]])
 
```
